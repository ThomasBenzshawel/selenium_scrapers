```python
def page_scrape(links_visited, driver):
    if len(links_visited) == 0:
        driver.get(url)
        driver.refresh()

    # Add the page source to the variable `content`.
    content = driver.page_source
    # Load the contents of the page, its source, into BeautifulSoup 
    # class, which analyzes the HTML as a nested data structure and allows to select
    soup = BeautifulSoup(content, 'html.parser')

    page_results = ""
    
    all_bubbles = soup.find_all(attrs={'class': 'bubble'})
    for element in all_bubbles:
        
        passage = ""
        if element is not None:
            if element.text != "":
                passage = passage + element.text + "\n"

#         speaker = element.find("a")
#         if speaker is not None:
#             name = speaker.text
#         else:
#             name = ""

#         info = element.find_all('p')

#         passage = ""
#         if info is not None:
#             for line in info:
# #                 for doing character atribution 
#                 passage = passage + name + ": " + line.text

#                 passage = passage + line.text



        if len(passage) > 1:
            page_results = page_results + passage + "\n"

    if len(page_results) > 1:    
        return page_results
```

```python
driver = webdriver.Firefox(options=opts)
driver.maximize_window()
url = "https://ffxiv.gamerescape.com/wiki/Category:Loremonger_Transcript"
driver.get(url)
driver.refresh()

links_visited = []
results = []
pages = 4769
num_scraped_already = 0

links = driver.find_elements("partial link text", "Loremonger:")
links_seen = len(links)
i_page = 0


print("------ STARTING FIRST PAGE ------")
for i in range(pages):
    if i != 0 and ((i_page) % links_seen == 0):
        print("------ NEXT PAGE ------")
        driver.find_elements("link text", "next page")[0].click()
        driver.refresh()
        links = driver.find_elements("partial link text", "Loremonger:")
        links_visited = []
        links_seen = len(links)
        i_page = 0

    if i >= num_scraped_already and i_page<len(links):
        link = links[i_page]
        print(i_page + 1)
        
        if link.text not in links_visited:
            print("Scraping", link.text, "...")
            links_visited.append(link.text)
            pre_click = driver.current_url
            link.click()
            driver.refresh()

#           results.append(page_scrape(links_visited, driver))

            page = page_scrape(links_visited, driver)

            with open('ff14_final_scrape.txt', 'a') as f:
                if page is not None:
                    f.write(page)
            
#             driver.refresh()
            current = driver.current_url
            print(current)
            if pre_click != current:
                driver.back()
                driver.refresh()
            print(driver.current_url)

            links = driver.find_elements("partial link text", "Loremonger:")
    i_page = i_page + 1
```
