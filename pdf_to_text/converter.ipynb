```python
# importing required modules 
from pypdf import PdfReader
import os 
```

```python
folder = "C:\\Users\\benzshawelt\\Downloads\\Convert"
save_folder = "C:\\Users\\benzshawelt\\Downloads\\Convert\\Save"

if not os.path.exists(save_folder):
    os.makedirs(save_folder)

# for each file in the folder
for file in os.listdir(folder):
    # if the file is a pdf
    if file.endswith(".pdf"):
        # open the pdf
        pdf = PdfReader(folder + "\\" + file)
        # get the number of pages
        num_pages = len(pdf.pages)
        # print the number of pages
        print(file + " has " + str(num_pages) + " pages")

        # initialize the text variable
        text_save = ""

        # get the text from each page
        for i in range(num_pages):
            #skip first page since it is the cover page
            if i != 0:
                text = pdf.pages[i].extract_text()
                # Save the text for each page so we have all text from the pdf
                text_save = text_save + text
        
        #clean out all text that is "Downloaded by Thomas Benz (thomas.benz.mc20@gmail.com)"
        text_save = text_save.replace("Downloaded by Thomas Benz (thomas.benz.mc20@gmail.com)", "")
        #Remove lOMoARcPSD|39146510
        text_save = text_save.replace("lOMoARcPSD|39146510", "")
        #remove all non ascii characters
        text_save = ''.join([i if ord(i) < 128 else ' ' for i in text_save])

        if text.strip() == "":
            print("No text found in " + file)
            continue

        # Save the text to a file
        with open(save_folder + "\\" + file + ".txt", "w", encoding="utf-8-sig") as f:
            f.write(text_save)
            
```
